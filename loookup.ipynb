```python
import os
import zipfile

# Base folder containing ZIP files
base_path = "/content"

# Loop through all ZIP files
for filename in os.listdir(base_path):
    if filename.lower().endswith(".zip"):
        zip_path = os.path.join(base_path, filename)
        print(f"‚è≥ Extracting '{filename}' ‚Üí '{base_path}'")
        # Create a new folder with the same name as the zip file (without .zip)
        extract_folder = os.path.join(base_path, os.path.splitext(filename)[0])

        # Ensure the folder exists
        os.makedirs(extract_folder, exist_ok=True)

        # Extract files there
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_folder)

        print(f"‚úÖ Extracted '{filename}' ‚Üí '{extract_folder}'")

print("\nüéâ All ZIP files extracted into their own folders successfully!")
```

```python
import os
import pandas as pd
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet

# Base path
base_path = "/content"
output_pdf = "/content/all_csv_data.pdf"

# Find all CSV files recursively
csv_files = []
for root, _, files in os.walk(base_path):
    for file in files:
        if file.lower().endswith(".csv") and not file.startswith("._"):
            csv_files.append(os.path.join(root, file))

# Setup PDF
styles = getSampleStyleSheet()
doc = SimpleDocTemplate(output_pdf, pagesize=A4)
elements = []

# Add data from each CSV
for path in csv_files:
    try:
        df = pd.read_csv(path)

        # Title
        elements.append(Paragraph(f"üìÇ File: {path}", styles['Heading3']))
        elements.append(Spacer(1, 10))

        # Add dataframe (only first 10 rows to keep it readable)
        data = [df.columns.tolist()] + df.head(10).values.tolist()
        table = Table(data, repeatRows=1)

        # Style table
        table.setStyle([
            ('BACKGROUND', (0, 0), (-1, 0), colors.lightgrey),
            ('GRID', (0, 0), (-1, -1), 0.25, colors.black),
            ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
            ('FONTSIZE', (0, 0), (-1, -1), 8),
        ])

        elements.append(table)
        elements.append(Spacer(1, 20))

        # Add summary
        elements.append(Paragraph(f"Rows: {len(df)}, Columns: {len(df.columns)}", styles['Normal']))
        elements.append(Spacer(1, 30))

    except Exception as e:
        elements.append(Paragraph(f"‚ö†Ô∏è Could not read {path}: {e}", styles['Normal']))
        elements.append(Spacer(1, 20))

# Build PDF
doc.build(elements)
print(f"‚úÖ PDF created successfully: {output_pdf}")
```
